```python
# Section 4 One-to-two Decay Modeling

import random
import math
import numpy as np
import matplotlib.pyplot as plt


coordinates = []

# First two arguments are starting point's coordinates, the last two are starting momentum vector's entries
def split(x, y, i, j):
    p1 = [x, y]
    p1mag = ((i**2) + (j**2))**0.5
    p1angle = math.atan( j / i )
    p2mag = random.random() * p1mag
    a = p1angle + (random.random() * math.pi / 2)
    
    p2_vector_x = p2mag * math.cos(a)
    p2_vector_y = p2mag * math.sin(a)
    p2 = [p2_vector_x + x ,p2_vector_y + y ,   p2_vector_x, p2_vector_y ]
    p3 = [ i - p2_vector_x + x,  j - p2_vector_y + y, i - p2_vector_x, j - p2_vector_y ]
    
    
    coordinates.append([p1[0], p1[1], p2[0], p2[1]])
    coordinates.append([p1[0], p1[1], p3[0], p3[1]])
   
    if p2mag < 0.2:
        return
    else:
        split(p2[0] , p2[1], p2[2], p3[3])
        split(p3[0], p3[1], p3[2], p3[3])
        return
        
    



split(1,0, 1, 0 )




for vector in coordinates:
    plt.plot([vector[0], vector[2]], [vector[1], vector[3]])
    
print ('you got the plot!')
```
